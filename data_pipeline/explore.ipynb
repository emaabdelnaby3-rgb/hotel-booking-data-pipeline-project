```python
import pandas as pd

df = pd.read_csv("../data/hotel_bookings.csv")
print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

# Cleaning and Data Quality

### Completeness

```python
print(df.isnull().sum())
```

###### agent                              16340
###### company                           112593
###### country                              488
###### children                               4

```python
# children ---> 4 FILL WITH ZERO CUZ SURE LW MTZAKRSH AKED NO KIDS
df["children"] = df["children"].fillna(0)

print(df.isnull().sum())
```

```python
# country  ------->  488  i will fill it with Unknown to save the row
df["country"] = df["country"].fillna("Unknown")
print(df.isnull().sum())
```

```python
# agent --------------->16340
df["agent"] = df["agent"].fillna(0)
print(df.isnull().sum())
```

```python
# company -------------> 112593 i will drop this columns cuz so much values are missing
df = df.drop(columns=["company"])
```

```python
print(df.isnull().sum())
```

### Uniqueness

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

### Validity

```python
# delete the negative values in price of adr
# just show
print(df[df["adr"] < 0])

# edit it
df = df[df["adr"] >= 0]
# check
print(df[df["adr"] < 0])

print(df["adr"].describe())
# thier is outliar the max is 5400 w en range from 70 to 134
```

```python
df[df["adr"] > 1000]
```

```python
# delete it
df = df[df["adr"] < 1000]
```

### Consistency

```python
# show length of booking guests
print(len(df[df["adults"] + df["children"] + df["babies"] > 0]))
```

```python
# remove bookings with zero guests
df = df[df["adults"] + df["children"] + df["babies"] > 0]
```

```python
# check
print(len(df[df["adults"] + df["children"] + df["babies"] == 0]))
```

##### children traveling without adults؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟؟

```python
# check bookings where children exist but no adults
len(df[(df["adults"] == 0) & (df["children"] > 0)])
```

```python
print(df[(df["adults"] == 0) & (df["children"] > 0)])
# save every row but el condition by match m3ah
df = df[~((df["adults"] == 0) & (df["children"] > 0))]
```

### Data Type Quality

```python
df["reservation_status_date"] = pd.to_datetime(df["reservation_status_date"])
```

### Feature Engineering

```python
# total_nights
df["total_nights"] = df["stays_in_week_nights"] + df["stays_in_weekend_nights"]
# total_guests
df["total_guests"] = df["adults"] + df["children"] + df["babies"]
```

```python
# save to silver layer
df.to_csv("../data/silver_hotel_bookings.csv", index=False)

print("el Silver Layer Created")
```
