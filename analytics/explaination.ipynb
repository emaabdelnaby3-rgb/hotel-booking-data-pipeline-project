```python
import pandas as pd

df = pd.read_csv("../data/silver_hotel_bookings.csv")

print(df.shape)
```

### Cancellation Rate

```python
cancelled_req_rate = df["is_canceled"].mean() * 100

print(f"Cancellation Rate: {round(cancelled_req_rate, 2)}")
```

### Average Price per Month

```python
# gropby bs 3mlt round
avg_price_month = round(df.groupby("arrival_date_month")["adr"].mean(), 2)

print(avg_price_month)
```

### Bookings per Country

```python
bookings_country = df["country"].value_counts().head(10)

print(bookings_country)
```

### Hotel Type Distribution

```python
# City vs Resort
hotel_dist = df["hotel"].value_counts()

print(hotel_dist)
```

### Average Stay Nights

```python
avg_nights = df["total_nights"].mean()

print("Average Stay:", round(avg_nights, 2))
```

```python
avg_price_month.to_csv("../data/gold_avg_price_per_month.csv")

bookings_country.to_csv("../data/gold_bookings_per_country.csv")
```
